```python
coll_n_weather['Deaths & Injuries'] = coll_n_weather['NUMBER OF PERSONS KILLED'] + coll_n_weather['NUMBER OF PERSONS INJURED']
```

```python
coll_n_weather.columns
```

```python
coll_n_weather['Deaths & Injuries'].sum()
```

```python
coll_n_weather['DATE'] = pd.to_datetime(coll_n_weather['DATE'])
```

```python
#create collsions column for sum of collisions each day.
coll_per_day_gp = coll_n_weather.groupby(['DATE'])
coll_per_day = coll_per_day_gp['DATE'].count()

coll_per_day = pd.DataFrame(coll_per_day)

coll_per_day = coll_per_day.rename(columns={"DATE": 'Daily Collision'})
coll_per_day.reset_index()

coll_per_day.head()
```
